```python
# Name the last location of Annie, from most recent episode.
characters_df.loc[16,'location']['name']
```

```python
# List the names of all episodes in which both Jerry Smith and Summer Smith appeared.
jerry=characters_df.loc[4,'episode']
summer=characters_df.loc[2,'episode']
both=(set(jerry).intersection(set(summer)))
both=list(both)
for i in both:
    x=requests.get(i)
    print(x.json()['name'])
```

```python
# List top 5 characters with maximum number of episodes. Start from the highest.
def length(i):
    return len(i)
characters_df['length']=characters_df['episode'].apply(length)
a=characters_df.sort_values(by='length',ascending=False)[['name','length']]
a.nlargest(5,'length')
```

```python
# What is the name of the episode with maximum number of characters?
def length(i):
    return len(i)
episodes_df['number of characters']=episodes_df['characters'].apply(length)
a=episodes_df.nlargest(1,'number of characters')
a[['name','number of characters']]
```
